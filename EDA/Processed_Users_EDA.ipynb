## Cleanup gross amount

```python
#remove dollar sign
ratings_df['total_gross'] = ratings_df['total_gross'].str.split('$', 1).str[1]
ratings_df['inflation_adjusted_gross'] = ratings_df['inflation_adjusted_gross'].str.split('$', 1).str[1]

#remove comma
ratings_df['total_gross'] = ratings_df['total_gross'].str.replace(',','')
ratings_df['inflation_adjusted_gross'] = ratings_df['inflation_adjusted_gross'].str.replace(',','')
ratings_df.head()
```

```python
ratings_df['total_gross'] = ratings_df['total_gross'].astype(float)
ratings_df['inflation_adjusted_gross'] = ratings_df['inflation_adjusted_gross'].astype(float)
```
