```python
# Import Dependencies
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#load in data set

tmdb_5000_movies = "Resources/tmdb_5000_movies.csv"
tmdb_5000_credits = "Resources/tmdb_5000_credits.csv"


#read datasets into DataFrames

tmdb_movies_df = pd.read_csv(tmdb_5000_movies)
tmdb_credits_df = pd.read_csv(tmdb_5000_credits)

```

```python

#relabel columns on tmdb_credits_df (movie_id = id) in order to merge on 'id'
tmdb_credits_df.columns = ['id', 'title', 'cast', 'crew']
tmdb_credits_df

```

```python
new_tmdb_movies_df = tmdb_movies_df.merge(tmdb_credits_df, on='id')
```

```python
new_tmdb_movies_df.head()
```

```python
#Evaluating Features

from ast import literal_eval
features = ["cast", "crew", "keywords", "genres"]
for feature in features:
    new_tmdb_movies_df[feature] = new_tmdb_movies_df[feature].apply(literal_eval)
new_tmdb_movies_df[features].head(10)
```

```python
# Function to extract Director of a movie
def get_director(x):
    for i in x:
        if i["job"] == "Director":
            return i["name"]
    return np.nan
```

```python
#Returns ranked list at least top 3

def get_list(x):
    if isinstance(x, list):
        names = [i["name"] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return []
```

```python
new_tmdb_movies_df["director"] = new_tmdb_movies_df["crew"].apply(get_director)
features = ["cast", "keywords", "genres"]
for feature in features:
   new_tmdb_movies_df[feature] = new_tmdb_movies_df[feature].apply(get_list)
```

```python
new_tmdb_movies_df.head()
```

```python
#Data clean up with a function

def clean_data(row):
    if isinstance(row, list):
        return [str.lower(i.replace(" ", "")) for i in row]
    else:
        if isinstance(row, str):
            return str.lower(row.replace(" ", ""))
        else:
            return ""

features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
   new_tmdb_movies_df[feature] = new_tmdb_movies_df[feature].apply(clean_data)
```

```python
# Define a function to scrape all metadata information into a vectorizer
def scrape(features):
    return ' '.join(features['keywords']) + ' ' + ' '.join(features['cast']) + ' ' + features['director'] + ' ' + ' '.join(features['genres'])
new_tmdb_movies_df["scrape"] = new_tmdb_movies_df.apply(scrape, axis=1)
print(new_tmdb_movies_df["scrape"].head())
```

```python
count_vectorizer = CountVectorizer(stop_words="english")
count_matrix = count_vectorizer.fit_transform(new_tmdb_movies_df["scrape"])
print(count_matrix.shape)

cosine_sim2 = cosine_similarity(count_matrix, count_matrix) 
print(cosine_sim2.shape)

new_tmdb_movies_df = new_tmdb_movies_df.reset_index()
indices = pd.Series(new_tmdb_movies_df.index, index=new_tmdb_movies_df['title'])
```

```python
new_tmdb_movies_df
```

```python
indices = pd.Series(new_tmdb_movies_df.index, index=new_tmdb_movies_df["original_title"]).drop_duplicates()
print(indices.head())
```

```python

def get_recommendations(original_title, cosine_sim):
    #Get index of movie
    idx = indices[original_title]
    #Get list of similarity scores and enumerate them
    similarity_scores = list(enumerate(cosine_sim[idx]))
    similarity_scores= sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similarity_scores= similarity_scores[1:8]

    # (a, b) where a is id of movie, b is similarity_scores
    movies_indices = [ind[0] for ind in similarity_scores]
    movies = new_tmdb_movies_df["original_title"].iloc[movies_indices]
    return movies


```

```python
print("Recommendations for John Carter")
print(get_recommendations("John Carter", cosine_sim2))


```
