## Here we are trying to match IMDB titles to Disney movie titles so we can query Movie Lens successfully

In this case, IMDB titles take precendence as over Disney titles as ML titles matches precisely to IMDB titles

### Load the FINAL Disney movies & IMDB movies

```python
#load IMDB movies dataset
imdb_df = pd.read_csv('input/IMDB_movies.csv')
imdb_df['Title'] = imdb_df['Title'].apply(str)
imdb_df['Title'] = imdb_df['Title'].str.replace('.','')
imdb_df['lower_title'] = imdb_df['Title'].str.lower()
imdb_df.head()
```

```python
imdb_df.dtypes
```

```python
#load disney movies dataset with gross $
disney_df = pd.read_csv('input/FINAL_2021_disney_movies_total_gross.csv', encoding='utf8')
disney_df.head()
```

```python
#load name list that has IMDB name vs Disney name

name_change = pd.read_csv('input/final_name_change.csv')
name_change.head()
```

```python
#Replace in disney dataset, the new names from IMDB final_name_change

for i in disney_df.index:
    for j in name_change.index:
        if disney_df['movie_title'][i] == name_change['movie_title'][j]:
            print("Change first movie title to second movie title")
            print(disney_df['movie_title'][i])
            print(name_change['Title'][j])
            disney_df['movie_title'][i] = name_change['Title'][j]
            print("--------------------------")
```
